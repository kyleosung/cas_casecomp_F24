```python
import pandas as pd

df = pd.read_csv('data/data_cleaned_twice.csv', index_col=0)
# df = df.drop(columns=['student_id', 'name', 'is_female'])
df.head()
```

```python
average_claim_amounts = df.groupby('class')['amount'].mean()
average_claim_amounts
```

```python
import matplotlib.pyplot as plt
plt.plot(average_claim_amounts)
plt.xlabel("Class")
plt.ylabel("Average Claim Amount")
plt.title("Personal Property: Average Claim Amount")
plt.show()
```
