```python
import pandas as pd

df = pd.read_csv('data/data_cleaned_twice.csv', index_col=0)
df = df.drop(columns=['student_id', 'name', 'is_female'])
df.head()
```

```python
X = df['class']
y = df['amount']

import matplotlib.pyplot as plt

plt.scatter(X, y)
plt.xlabel('Class')
plt.ylabel('Claim Amount')
plt.title('Claim Amount vs Class')
plt.show()
```

```python
for x_label in df.columns:
    for y_label in ['risk_tier', 'amount']:
        
        if x_label == y_label:
            continue
        
        X = df[x_label]
        y = df[y_label]

        import matplotlib.pyplot as plt

        plt.scatter(X, y)
        plt.xlabel(f'{x_label}')
        plt.ylabel(f'{y_label}')
        plt.title(f'Personal Property: {y_label} vs {x_label}')
        plt.savefig(f'img_PP_{x_label}_{y_label}.png')
        plt.show()
```
