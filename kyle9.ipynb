```python
import pandas as pd

df = pd.read_csv('data/data_cleaned_twice.csv', index_col=0)
df = df.drop(columns=['student_id', 'name', 'is_female'])
df.head()
```

```python
X = df['class']
y = df['amount']
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, y)
plt.xlabel('Class')
plt.ylabel('Claim Amount')
plt.title('Claim Amount vs Class')
plt.show()
```
