```python
import pandas as pd
```

```python
df = pd.read_csv("./data/PredictiveModelingDataset.csv")
df.head()
```

```python
binary_columns = ['greek', 'off_campus', 'gender', 'sprinklered', 'holdout']
df[binary_columns] = df[binary_columns].apply(lambda x: x.map({'Non-greek': 0, 'Greek': 1, 'On campus': 0, 'Off campus': 1, 'Male': 0, 'Female': 1, False: 0, True: 1}))
df

df.rename(columns={'greek': 'is_greek', 'off_campus': 'is_on_campus', 'gender': 'is_female', 'sprinklered': 'is_sprinklered'}, inplace=True)
df
```

```python
df['class'].unique()
class_mapping = {'Freshman': 1, 'Sophomore': 2, 'Junior': 3, 'Senior': 4, 'Grad student': 5}
df['class'] = df['class'].map(class_mapping)
df['class'].unique()
```

```python
df.to_csv('./data/data_cleaned.csv')
```
