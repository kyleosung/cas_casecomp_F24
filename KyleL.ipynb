```python
import pandas as pd
import matplotlib.pyplot as plt

for dataset in ['AdditionalLivingExpense', 'Guest_medical', 'liability']:
    df = pd.read_csv(f'data/data_cleaned_{dataset}.csv', index_col=0)
    # df = df.drop(columns=['student_id', 'name', 'is_female'])
    df.head()

    average_claim_amounts = df.groupby('class')['amount'].mean()
    average_claim_amounts


    plt.plot(average_claim_amounts)
    plt.xlabel("Class")
    plt.ylabel("Average Claim Amount")
    plt.title(f"{dataset}: Average Claim Amount")
    plt.show()
```

```python
for dataset in ['AdditionalLivingExpense', 'Guest_medical', 'liability']:
    for y_label in ['risk_tier', 'amount']:
        df = pd.read_csv(f'data/data_cleaned_{dataset}.csv', index_col=0)
        # df = df.drop(columns=['student_id', 'name', 'is_female'])
        df.head()

        X = df['class']
        y = df[y_label]

        plt.scatter(X, y)
        plt.xlabel("Class")
        plt.ylabel(f"{y_label}")
        plt.title(f"{dataset}: {y_label} by Class")
        plt.show()
```
